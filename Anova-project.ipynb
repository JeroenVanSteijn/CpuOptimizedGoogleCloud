```python
rp.summary_cont(df.groupby(['Setup', 'Batch_Size', 'Max_Epochs', 'Learning_Rate', 'Memory_Size', 'Model']))['Training_Time']
```

```python
model = ols('Training_Time ~ C(Setup)*C(Batch_Size)*C(Max_Epochs)*C(Learning_Rate)*C(Memory_Size)*C(Model)', df).fit()

# Seeing if the overall model is significant
print(f"Overall model F({model.df_model: .0f},{model.df_resid: .0f}) = {model.fvalue: .3f}, p = {model.f_pvalue: .4f}")
```

```python
from sklearn import preprocessing
#plt.scatter(df['Training_Time'], df2['Accuracy'])
le = preprocessing.LabelEncoder()


# Fit the setups
le.fit(['standard', 'hybrid', 'optimized'])
df['Setup_label'] = le.transform(df['Setup'])
print(le.classes_)

# Fit the models
le.fit(['lenet5', 'bi-rnn'])
df['Model_label'] = le.transform(df['Model'])
print(le.classes_)

X = df[['Batch_Size', 'Max_Epochs', 'Setup_label', 'Model_label']]
Y = df['Training_Time']


regr = linear_model.LinearRegression()
regr.fit(X,Y)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)


new_batch_size = 128
new_epochs = 20
new_model = 0

regr.predict([[new_batch_size, new_epochs, 2, new_model], 
              [new_batch_size, new_epochs, 0, new_model],
              [new_batch_size, new_epochs, 1, new_model]])
```

```python
pd.set_option('display.max_rows', 1500)
res = sm.stats.anova_lm(model, typ= 2)
res
```
